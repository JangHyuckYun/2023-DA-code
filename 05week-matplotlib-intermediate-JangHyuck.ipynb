```python
import matplotlib.pyplot as plt
import numpy as np

ax1 = plt.subplot(2,1,1)
plt.plot(np.random.randint(1, 10, 10), 'yo-')
plt.title('a tale of 2 subplots')
plt.ylabel('damped osciliation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(np.arange(10), np.arange(10) * 2, 'r.-')
plt.title('a tale of 2 subplots')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)
axes[0, 0].plot([1,2,3,4])
axes[0, 0].set_title("Subplot 1")

axes[0, 1].plot([4,3,2,1])
axes[0, 1].set_title("Subplot 2")

axes[1, 0].hist([1,2,2,3,3,3,4,4,4,4], bins=5)
axes[1, 0].set_title("Subplot 3")

axes[1, 1].plot([1,2,3,4,], [4,3,2,1])
axes[1, 1].set_title("Subplot 4")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(range(-20, 80), np.random.randn(100).cumsum(), 'r--')
plt.title('한글 점검');
```

```python
fig = plt.figure(figsize=(15, 12))
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'),
figsize=(15, 12))
graph_title = '테스트'
ax.set_title(f'{graph_title}')
ax.set_xlabel('주파수 (Hz)', labelpad=20)
ax.set_ylabel('초 (s)', labelpad=20)
ax.set_zlabel('강도/주파수 (dB/Hz)', labelpad=20)
ax.view_init(20, 20)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize = (8, 6))

for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='c', alpha=1)
        
plt.subplots_adjust(wspace=0, hspace=0.2)
```
